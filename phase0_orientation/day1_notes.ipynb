## Step by step instructions :
- Day 1 has three concrete steps: (1) API key, (2) environment, (3) first call. 

### Step 0: Check your Python version 
- Open your terminal and type : ``python3 --version`` . Proceed if you see something like Python 3.11.x or higher.

### Step 1: Get your API key 
- Open a browser and go to https://console.anthropic.com
- Sign up for an account (or log in if you have one). Anthropic will verify your email.
- Once logged in, buy $5 in credits (the minimum). That's more than enough for all of Phase 0 and 
then some — you'll spend maybe $0.50 across all six days.
- After that look for "API Keys" (under "Account" or "Settings").
- Click "Create API Key" and name it ``evalkit-supritis``.
- Copy and save the API Key.It starts with sk-... and is long. Store it somewhere safe (a password manager, a temp file, somewhere you can paste from). 

### Step 2 : Set the API key as an environment variable
- The goal: your code can read ANTHROPIC_API_KEY without you typing it or pasting it in code.
- On Mac/Linux terminal type : 
    - nano ~/.zshrc
- This opens a text editor. Paste this line:    
    - export ANTHROPIC_API_KEY=sk-[paste-your-actual-key-here]
- Then :  
    - Press Ctrl + O (the letter O, not zero)
    - Press Enter to confirm the filename
    - Press Ctrl + X to exit   
- Then reload it : source ~/.zshrc
- Verify : echo $ANTHROPIC_API_KEY   

### Step 3 :  Create the project folder & Python environment
- Open a terminal in a clean spot where you keep projects and move into that.
- Now create a Python virtual environment (this isolates your dependencies so they don't mess with your system Python)
    - python3 -m venv venv
- This makes a folder called venv. Activate it :
    - source venv/bin/activate

### Step 3 : Initialize Git and Create .gitignore
- Under your (venv) and required folder type : ``git init``
    - Expected O/P - Initialized empty Git repository in /Users/supritis/Documents/AI_Learnings/AI_EvalKit/.git/
- Now set the main branch name: ``git branch -M main``
- Next create a .gitignore file so you don't accidentally commit your API key or other secrets :
    - Run in terminal : 
```
cat > .gitignore << 'EOF'
# Python virtual environment
venv/
.venv/

# API keys and environment variables
.env
.env.*

# Python-generated files
__pycache__/
*.py[cod]

# Jupyter-generated files
.ipynb_checkpoints/

# macOS-generated files
.DS_Store

# VS Code local settings
.vscode/
EOF

```

- Now run : ``git status`` . Output should only have - 

```
On branch main

No commits yet

Untracked files:
  .gitignore
  Phase0/

```  

### Step 4 : Install the Anthropic SDK 
- Upgrade pip (the package installer) if needed :
    - ``pip install --upgrade pip``
- Ensure you are under your Folder : AI_EvalKit and venv for that. Then type :
    - ``pip install anthropic python-dotenv``
        - What above command does? - Installs the Claude SDK and then the tool that securely loads this project’s API key. ``python-dotenv`` is included because we plan to store the API key in a .env file later.
    - Check if both installed - ``python3 -c "import anthropic; import dotenv; print('Both good')" ``

### Step 5 : Write your first script
- Create a file called ``day1_hello_model.py`` under Phase0.
- Add code (see file for code details)

### Step 6 : Run your first call
- Run the file - ``day1_hello_model.py`` and check the o/p in terminal

### Step 7 : Commit your work and Push to Git Repo
- First stage and save changes locally :
    - ``git status``
    - ``git add . ``
    - ``git commit -m "This is my First ....."`` ---> saves your work locally
    - ``git log``
- Next create a GitHub Repository and then push code from local to GitHub
    - Created : https://github.com/supritism/AI-EvalKit
- Then connect your local project to GitHub
    - ``git remote add origin https://github.com/supritism/AI-EvalKit.git``
- Check that connection is correct
    - ``git remote -v``
- Push your local commit to GitHub
    - ``git push -u origin main``
    - Note : The first time we push from local to Git repo, Git needs the full destination then future uploads just ``git push`` works.
    - Note : ``-u`` remembers that local ``main`` is connected to ``origin/main``

### Day 2 
- Understanding how many tokens a prompt uses before I send it. The gateway to cost control and smart eval design.

### Questions for next phase
- How do I count tokens *before* making the call?
- What makes a good "golden set" of test cases?
